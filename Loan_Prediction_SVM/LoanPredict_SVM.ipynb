```python
#import the necessary libraries
import numpy as np
import pandas as pd
 

#create a dataframe from an uploaded excel file
dataFrame = pd.read_excel("/Users/tarieretimitimi/Documents/Programming/Python/SVM_Project/Loan_Data.xlsx")
X = dataFrame.iloc[:,2:5] # Independent variable: FICO score, Installments and Interest rate
y = dataFrame.iloc[:, 5]  # Dependent variable: Not fully paid

#split the data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

#scale the data for faster performance
from sklearn.preprocessing import StandardScaler
STsc = StandardScaler()
X_train = STsc.fit_transform(X_train)
X_test = STsc.transform(X_test)

#Create the support vector machine
from sklearn.svm import SVC
OurSVM = SVC(kernel = 'poly', random_state = 0)
OurSVM.fit(X_train, y_train)

#get the predicted amounts using the values in X_test
y_pred = OurSVM.predict(X_test)


#Create a confusion matrix to see how accurate our model is
from sklearn.metrics import confusion_matrix
OurCM = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score
print ("The Model's Accuracy is : ", accuracy_score(y_test, y_pred))
print ("The Confusion Matrix is:" )
print(OurCM)

print("\n")

print ("The Confusion Matrix shows:" )
print ("TRUE NEGATIVE  |  FALSE POSITIVE")
print ("--------------------------------" )
print ("FALSE NEGATIVE |  TRUE POSITIVE" )

print("\n")

#compare the real values to those predicted by the support vector machine
df = pd.DataFrame({'Actual Values':y_test, 'Predicted Values':y_pred})
print(df)

#Use the model to classify whether there's a recession for
#three new values for inflation unemployment and GNP

#Use the normalize method to scale only one group of values
from sklearn import preprocessing
data = np.array([80, 0.075, 725])
norm_data = preprocessing.normalize([data])

y_pred = OurSVM.predict(norm_data)

print("\n")
print("Using the values of $80 for the installment amount, 0.075 for the interest rate and 725 for the fico score")
print("the prediction for if they will pay back the loan( where 1 is no, 0 is yes):")
print (y_pred)
```

The SVM model demonstrated a 90% accuracy in predicting loan repayment status. The confusion matrix revealed that the model was better at identifying loans likely to be fully paid than those likely to default and this is likely a reflection of the data used.  The confusion matrix provides a detailed breakdown of the model's performance. 
•	True Positive (TP): The model correctly predicted 27 instances where the loan was not fully paid
•	False Positive (FP): The model incorrectly predicted 2 instances where the loan was fully paid
•	False Negative (FN): The model incorrectly predicted 1 instance where the loan was not fully paid 
•	True Negative (TN): The model correctly predicted 0 instances where the loan was fully paid
When predicting the outcome for a new loan with an $80 installment, 7.5% interest, and a 725 FICO score, the model predicted that it would be fully paid.  This suggests that individuals with these financial characteristics are more likely to pay their loan fully.
